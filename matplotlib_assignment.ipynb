##1. Draw a line plot of given data below with the following details:

#####Add grid to the plot
#####Add the title as "Visualizing dependency of x_label and y_label"
#####Add xlabel as x_label
#####Add ylabel as y_label
#####Add legend consisting of x_label and y_label

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = [2,4,6,8,10,11,11.5,11.7]
b = [1,1.5,2,2.5,3,3.5,4,4.5]

plt.plot(a,b,marker ='o')

plt.title('Visualizing dependency of x_label and y_label',fontsize=15)
plt.xlabel('x_label')
plt.ylabel('y_label')

plt.legend(['a-b'])
plt.grid()
plt.show()
```

#2. Draw a scatter plot of given data below with the following details:

#####Add the title as "Visualizing dependency of x_label and y_label"

#3. Draw the bar graph with the given data and add the following details:

#6. Use subplot function to plot given sine and cosine functions

#####First plot them side side by horizontally i.e., two coloumns and a single row
#####Second plot them up and down i.e., single coloumn and two rows

```python
x = np.arange(0, 3 * np.pi, 0.1) 
y_sin = np.sin(x) 
y_cos = np.cos(x) 

plt.subplot(1,2,1)
plt.plot(x,y_sin)
plt.title('Sine')
plt.grid()
plt.subplot(1,2,2)
plt.plot(x,y_cos)
plt.title('Cosine')

plt.grid()
plt.show()
print(' ')#skipping a line
#VERTICAL
plt.subplot(2,1,1)
plt.plot(x,y_sin)
plt.title('Sine')
plt.grid()
plt.subplot(2,1,2)
plt.plot(x,y_cos)
plt.title('Cosine')

plt.grid()
plt.show()
```
